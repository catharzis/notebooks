# Regression

## Step 0

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples= 10, n_features=3, random_state=45)
```

```python
 X
```

```python
y
```

```python
df = pd.DataFrame(X)
```

```python
df['y_true'] = y
```

```python
df['y_pred_0'] = df['y_true'].mean()
```

```python
df.sort_values(by=0)
```

## Step 1

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mean_absolute_error(df['y_true'], df['y_pred_0'])
```

```python
df
```

```python
df['residual_0'] = df['y_true'] - df['y_pred_0']
```

```python
df
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_1 = DecisionTreeRegressor(max_depth=1)
```

```python
tree_1.fit(df[[0, 1, 2]], df['residual_0'])
```

```python
df['tree_pred_1'] = tree_1.predict(df[[0, 1, 2]])
```

```python
df
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(tree_1)
```

```python
nu = 0.1
```

```python
df['y_pred_1'] = df['y_pred_0'] + nu*df['tree_pred_1']
```

```python
df
```

```python
mean_absolute_error(df['y_true'], df['y_pred_0'])
```

```python
mean_absolute_error(df['y_true'], df['y_pred_1'])
```

```python
n = 10
nu = 0.3
trees = []

df['y_pred'] = df['y_true'].mean()

for i in range(n):
  df['residual'] = df['y_true'] - df['y_pred']
  tree = DecisionTreeRegressor(max_depth=1)
  tree.fit(df[[0, 1, 2]], df['residual'])
  df['y_pred'] += nu*tree.predict(df[[0, 1, 2]])
  trees.append(tree)
  print(mean_absolute_error(df['y_true'],df['y_pred']))
```

# Classification

```python
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=10, n_features=3, n_redundant=0, random_state=45)
```

```python
df = pd.DataFrame(X)
df['y_true'] = y
```

```python
df
```

```python
df['y_pred'] = df['y_true'].mean()
```

```python
df
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(df['y_true'], df['y_pred'])
```

```python
n = 10
nu = 0.3
trees = []

df['y_pred'] = df['y_true'].mean()

for i in range(n):
  df['residual'] = df['y_true'] - df['y_pred']
  tree = DecisionTreeRegressor(max_depth=1)
  tree.fit(df[[0, 1, 2]], df['residual'])
  df['y_pred'] += nu*tree.predict(df[[0, 1, 2]])
  trees.append(tree)
  print(roc_auc_score(df['y_true'], df['y_pred']))
```
