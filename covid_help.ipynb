```python
import pandas as pd
```

```python
df = pd.read_excel('data.xlsx', sheet_name=1)
df.head(3)
```

```python
df.dtypes
```

## Убираем переносы строк

```python
# вариант 1
list_of_cols = []

for i in df.columns.tolist():
  list_of_cols.append(i.replace('\n', ''))

df.set_axis(list_of_cols, axis=1, inplace=True)
```

```python
df.columns
```

```python
# вариант 2
df = df.rename(columns=lambda x: x.replace('\n', ''))
```

## Сводные

```python
import numpy as np
```

```python
pivot = pd.pivot_table(df,
               index=['Марка вакцины', 'Страна', 'Дата поставки/новости'],
               values=['Вакцина (кол-во)'],
               aggfunc=[np.sum])
```

```python
pivot = pivot.sort_values(by='Дата поставки/новости')
pivot
```

## Сохранение

```python
writter = pd.ExcelWriter('pivot_1.xlsx')
pivot.to_excel(writter, 'Выгрузка')
writter.save()
```

## Домены

```python
ser1 = df['Источник 1']
ser2 = df['Источник 2']
ser3 = df['Источник 3']

col_1 = ser1.str.split('/', expand=True, n=3)[2]
col_2 = ser2.str.split('/', expand=True, n=3)[2]
col_3 = ser3.str.split('/', expand=True, n=3)[2]

df['domain_1'] = col_1
df['domain_2'] = col_2
df['domain_3'] = col_3
```
